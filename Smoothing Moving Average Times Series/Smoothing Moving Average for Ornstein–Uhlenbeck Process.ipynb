# Task 4 part 2  - Ornstein–Uhlenbeck Process with Smoothing Moving Average

### Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import csv
```

### The Ornstein–Uhlenbeck Process 
is defined by the following stochastic differential equation:

- dxt/dt = -axtdt + σ*Wdt

### Initialize t and insert 1000 random values of values from -1 to 1 and put values in front of w


```python
t1 = np.zeros(1000)
t2 = np.zeros(1000)
s1 = 2
s2 = 120
```

```python
w1 = np.zeros(1000)
w2 = np.zeros(1000)
a = np.zeros(1000)
for i in range(1000):
    w1[i] = round(random.uniform(-1.00,1.00),1)
    w2[i] = round(random.uniform(-1.00,1.00),1)
    a[i] = round(random.uniform(-1.00,1.00),1)
#a =0.9
```

### Save the t1,t2 and w values to a .csv file

```python
with open('OData.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['w1', 'w2', 'a'])
    for i in range(1000):
        writer.writerow([w1[i], w2[i], a[i]])
```

### Putting Moving Average window


```python
window_size = 15
```

### Applying the equation for the t and s = 30

- dxt/dt = -axtdt + σ*Wdt

### where a takes random values

```python
t1[0] = 1
for i in range(1,1000):
    if i != 0 :
        t1[i] = -a[i]*t1[i-1] + s1*w1[i]
```

### Creating new time-series with Moving Average for s = 30

```python
smoothed_t1 = np.convolve(t1, np.ones(window_size) / window_size, mode='valid')
```

### Plot of the results of the equation for s = 30 and their Moving Average

```python
plt.figure(figsize=(20, 12))
plt.plot(t1, label='Original t1')
plt.plot(smoothed_t1, label='Smoothed t1')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

### Calculation of corresponding difference with s = 30 

```python
diff1 = [t1[i] - t1[i-1] for i in range(1,len(t1))]
```

### Applying the equation for the t and s = 120

- dxt/dt = -axtdt + σ*Wdt

### where a takes random values

```python
t2[0] = 1
for i in range(1,1000):
    if i != 0 :
        t2[i] = -a[i]*t2[i-1] + s2*w2[i]
```

### Creating new time-series with Moving Average for s = 120

```python
smoothed_t2 = np.convolve(t2, np.ones(window_size) / window_size, mode='valid')
```

### Plot of the results of the equation for s = 120 and their Moving Average

```python
plt.figure(figsize=(20, 12))
plt.plot(t2, label='Original t2')
plt.plot(smoothed_t2, label='Smoothed t2')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

### Calculation of corresponding difference with s = 120

```python
diff2 = [t2[i] - t2[i-1] for i in range(1,len(t2))]
```

### Creating new time-series with Moving Average for s = 30 and s = 120 with their corresponding difference


```python
smoothed_diff1 = np.convolve(diff1, np.ones(window_size) / window_size, mode='valid')
smoothed_diff2 = np.convolve(diff2, np.ones(window_size) / window_size, mode='valid')
```

### Plot of the results of the equation for s = 30 and s = 120 with their Moving Average

```python
plt.figure(figsize=(20, 12))
plt.plot(diff1, label='Original diff1')
plt.plot(smoothed_diff1, label='Smoothed diff1')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 12))
plt.plot(diff2, label='Original diff2')
plt.plot(smoothed_diff2, label='Smoothed diff2')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
