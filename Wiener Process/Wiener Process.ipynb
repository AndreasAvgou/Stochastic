# Task 2 - Wiener Process


### Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import csv
from statsmodels.graphics.tsaplots import plot_acf
```

```python
t1 = np.zeros(1000)
t2 = np.zeros(1000)
w1 = np.zeros(1000)
w2 = np.zeros(1000)
s = 30
size = 1000
```

### Initialize t and insert 1000 random values of values from -1 to 1 and put values in front of w

```python
for i in range(1000):
    t1[i] = round(random.uniform(-1.00, 1.00), 1)
    t2[i] = round(random.uniform(-1.00, 1.00), 1)
    w1[i] = round(random.uniform(-1.00, 1.00), 1)
    w2[i] = round(random.uniform(-1.00, 1.00), 1)
```

### Save the t1,t2 and w values to a .csv file

```python
with open('WData.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['w'])
    for i in range(1000):
        writer.writerow([t1[i],t2[i],w1[i],w2[i]])
```

### Applying the equation for the w values 
- g(t)= g(t-1) + w(t) * dt

```python
t1[0] = 1
t2[0] = 1
for i in range(1,size):
    if i != 0 :
        t2[i] = t2[i-1] + s * np.sqrt(1 / size) * w2[i]
        t1[i] = t1[i-1] + s * np.sqrt(1 / size) * w1[i]
```

```python
plt.figure(figsize=(20, 12))
plt.plot(t1, label='Original t1')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 12))
plt.plot(t2, label='Original t2')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
diff1 = [t1[i] - t1[i-1] for i in range(1, len(t1))]
diff2 = [t2[i] - t2[i-1] for i in range(1, len(t2))]
```

```python
plt.figure(figsize=(20, 12))
plt.plot(diff1, label='Original diff1')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 12))
plt.plot(diff2, label='Original diff2')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
