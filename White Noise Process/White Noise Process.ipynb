# Task 1 - White Noise


### Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import csv
from statsmodels.graphics.tsaplots import plot_acf
```

```python
t1 = np.zeros(1000)
t2 = np.zeros(1000)
w = np.zeros(1000)
s1 = 30
s2 = 120
```

### Initialize t and insert 1000 random values of values from -1 to 1 and put values in front of w

```python
for i in range(1000):
    t1[i] = round(random.uniform(-1.00, 1.00), 1)
    t2[i] = round(random.uniform(-1.00, 1.00), 1)
    w[i] = round(random.uniform(-1.00, 1.00), 1)
```

### Save the t1,t2 and w values to a .csv file

```python
with open('WData.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['w'])
    for i in range(1000):
        writer.writerow([t1[i],t2[i],w[i]])
```

### Applying the equation for the w values and with a value of s = 30 and s = 12-
- g(t)= g(t-1) + Ïƒ * w(t) * dt

```python
t1[0] = 1
t2[0] = 1
for i in range(1,1000):
    if i != 0 :
        t1[i] = t1[i-1] + s1*w[i]
        t2[i] = t2[i-1] + s2*w[i]
```

```python
plt.plot(t1)
plt.title('t1 values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.plot(t2)
plt.title('t2 values')
plt.xlabel('Index')
plt.ylabel('Value')
```

```python
diff1 = [t1[i] - t1[i-1] for i in range(1, len(t1))]
diff2 = [t2[i] - t2[i-1] for i in range(1, len(t2))]
```

```python
plt.figure()
plt.plot(diff1)
plt.title('Differences between successive t1 values')
plt.xlabel('Index')
plt.ylabel('Value')
```

```python
plt.plot(diff2)
plt.title('Differences between successive t2 values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
