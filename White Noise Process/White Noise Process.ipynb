# Εργασία 1 - White Noise

### Εισαγωγή απαραίτητων βιβλιοθηκών

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import csv
from statsmodels.graphics.tsaplots import plot_acf
```

```python
t = np.zeros(1000)
a = 30
```

### Αρχικοποίηση του t και εισαγωγή 1000 τυχαίων τιμών τιμών από -1 έως 1 και τοποθέτηση τιμών μπροστά από το w


```python
w = np.empty(1000,dtype=float)
for i in range(1000):
    w[i] = round(random.uniform(-1.00,1.00),1)
```

### Αποθήκευση των τιμών w σε ένα αρχείο .csv

```python
with open('WData.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['w'])
    for i in range(1000):
        writer.writerow([w[i]])
```

### Εφαρμογή της εξίσωσης για τις w τιμές και με τιμή σ = 30 
- g(t)= g(t-1) + σ * w(t) * dt

```python
t[0] = 1
for i in range(1,1000):
    if i != 0 :
        t[i] = t[i-1] + a*w[i]
```

```python
plt.plot(t)
```

```python
diff = [t[i] - t[i-1] for i in range(1,len(t))]
```

```python
plt.plot(diff)
```
